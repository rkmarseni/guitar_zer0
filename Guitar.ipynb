```python
import numpy as np
notes=['a','a#','b','c','c#','d','d#','e','f','f#','g','g#']
note_traverse=np.arange(0,22)
```

```python
def notes_in_string(st):
    string_notes=[]
    for i in note_traverse:
        string_notes.append(notes[(notes.index(st)+i+1)%len(notes)])
    return string_notes

def major_scales(of_note):
    whole=2
    half=1
    location=notes.index(of_note)
    scale=[]
    for a in range(8):
        scale.append(notes[(location)%len(notes)])
        if (a==2) or (a==6):
            location= (location+half)
        else:
            location=(location+whole)%len(notes)
    return scale

def chord_notes(ch):
    scale=major_scales(ch)
    ch_notes=[scale[0],scale[2],scale[4]]
    return ch_notes

def fretboard():
    str1_e=notes_in_string('e')
    str2_b=notes_in_string('b')
    str3_g=notes_in_string('g')
    str4_d=notes_in_string('d')
    str5_a=notes_in_string('a')
    str6_e=notes_in_string('e')
    list_of_strings=[str1_e,str2_b,str3_g,str4_d,str5_a,str6_e]
    return list_of_strings
    
```

```python
list_of_strings=fretboard()
print(list_of_strings[0])
```

```python
def chord_chart(chord):
    color=['lime','tomato','sandybrown','orchid','deepskyblue']
    ch_notes=[notes.index(chord),(notes.index(chord)+4)%len(notes),(notes.index(chord)+7)%len(notes)]
    print('Chord = ',ch_notes)
    ch_notes_labels=[notes[ch_notes[0]],notes[ch_notes[1]],notes[ch_notes[2]]]
    print('Notes = ',ch_notes_labels)
    plt.figure(figsize=(16,5))
    for x,string in enumerate(list_of_strings):
        a=0
        for y, note in enumerate(string):
            plt.text(y+.5, x+2,note, fontsize=12)
            if(note in ch_notes_labels):
                plt.plot(y+.35, x+2, color=color[a%len(color)], marker='o', markersize=12)
            a=a+1
    plt.grid(True)
    plt.axis([0, 22, 0, 9])
    xlabels = np.arange(1,22,1)
    plt.xticks(xlabels,xlabels)
    ylabels=['','','E','B','g','D','A','E','']
    ynum=np.arange(0,9,1)
    plt.yticks(ynum,ylabels)
    plt.show()
```

```python
ch=input('Enter a chord : ')
chord_chart(ch)
```
